# Appendix B

```python
import random
import numpy
import matplotlib.pyplot as plt
```

```python
counts = numpy.array([[6,0,1],
                             [0,7,0],
                             [2,0,1]], dtype=float)

plt.imshow(counts, cmap='Reds', interpolation=None)
```

```python
probabilities = counts/8
probabilities
```

```python
r = numpy.random.uniform(size=(3,3))
diff = probabilities - r
img = numpy.ceil(diff)

plt.imshow(img, cmap='Greens', interpolation=None)
```

```python
f, axarr = plt.subplots(5, 8, figsize=(12,6))
for i in range(5):
  for j in range(8):
    r = numpy.random.uniform(size=(3,3))
    diff = probabilities - r
    img = numpy.ceil(diff)
    axarr[i,j].set_xticks([])
    axarr[i,j].set_yticks([])
    axarr[i,j].imshow(img, interpolation='none', cmap='Blues')
```

#Appendix D

```python
import pandas
import matplotlib.pyplot as plt
```

```python
x = [0.6]
y = [0.7]

lr= 0.1
```

```python
for i in range(499):
  x.append(x[i] + lr * y[i])
  y.append(y[i] - lr * x[i])
```

```python
df = pandas.DataFrame(data=list(zip(x,y)), columns=['x','y'])
df.plot(figsize=(20,6))
```

```python
df.plot.scatter('x','y', figsize=(10,10))
```
